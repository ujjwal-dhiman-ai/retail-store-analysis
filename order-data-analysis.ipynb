```python
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
# read data from file and handle null values

import pandas as pd

df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df.head(20)
```

```python
df['Ship Mode'].unique()
```

```python
# rename column names - make them in lower case and remove whitespace

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df.head()
```

```python
# dervive new columns - discount, sales price, profit

df['discount'] = df['discount_percent']*df['list_price']*0.01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df.head()
```

```python
df.dtypes
```

```python
# convert order date from object data type to date 

df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
df.dtypes
```

```python
df.head()
```

```python
# drop unnecessary columms

df.drop(columns=['list_price','cost_price', 'discount_percent'], inplace=True)
```
